```python
# Import Libraries
import os
import urllib.request
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import keras_hub
from PIL import Image
```

```python
# Load ViTImageClassifierPreprocessor (vit_base_patch16_224_imagenet preset)
preprocessor = keras_hub.models.ViTImageClassifierPreprocessor.from_preset(
    "vit_base_patch16_224_imagenet"
)
```

```python
# Load ViTImageClassifier (vit_base_patch16_224_imagenet preset)
model = keras_hub.models.ViTImageClassifier.from_preset(
    "vit_base_patch16_224_imagenet"
)
```

```python
# Load the images
with open("imagenet_classes.txt") as f:
    class_names = [line.strip() for line in f]

print("Total classes:", len(class_names))
```

```python
image_files = [
    "bulbul.jpg",
    "pug.jpg",
    "Siamese_cat.jpg",
    "stingray.jpg",
    "zebra.jpeg"
]

images = []

for file in image_files:
    img = Image.open(file).convert("RGB")
    img = img.resize((224, 224))   # make batch rectangular
    images.append(np.array(img, dtype=np.uint8))

images = np.array(images)
print(images.shape, images.dtype)
```

```python
# Predict classes
predictions = model.predict(images)
predicted_indices = np.argmax(predictions, axis=1)

for i, idx in enumerate(predicted_indices):
    print(image_files[i], ":", class_names[idx])
```
